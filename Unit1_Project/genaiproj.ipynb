```python
from transformers import pipeline
```

```python
sentiment_analyzer = pipeline("sentiment-analysis")
```

```python
review = "The product quality is excellent and delivery was fast."

result = sentiment_analyzer(review)
print(result)
```

```python
reviews = [
    "The product is amazing and works perfectly.",
    "Worst experience ever. Totally disappointed.",
    "Customer support was okay, nothing special.",
    "I love this! Will buy again.",
    "The item arrived damaged and late."
]

for review in reviews:
    result = sentiment_analyzer(review)[0]
    print(f"Review: {review}")
    print(f"Sentiment: {result['label']} (Confidence: {result['score']:.2f})\n")
```

```python
user_review = input("Enter a customer review: ")
result = sentiment_analyzer(user_review)[0]
print(f"Sentiment: {result['label']} (Confidence: {result['score']:.2f})")
```
